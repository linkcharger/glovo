```python
import numpy as np
import pandas as pd
import dask.dataframe as dd 
from pyaxis import pyaxis
from pprint import pprint
from glob import glob

from glob import glob
from sys import getsizeof
import warnings
# warnings.filterwarnings("ignore")


```

```python
ddf = dd.read_parquet('01_census/data/parquet/')
ddf.head()
```

```python
len(ddf)
```

```python
ddf = ddf[~ddf['Municipality code or size'].isin([991.0, 992.0, 993.0, 994.0])]
ddf.head()
```

```python
len(ddf)
```

```python
ddf['locationCode'] = ddf.apply(lambda postcode: f'{int(postcode["province code"]):02}{int(postcode["Municipality code or size"]):03}', axis=1).astype(str)
ddf.head()
```

```python
locationDF = pd.read_excel('01_census/data/provincial and municipal codes.xlsx')

def locationLookup(locationCode):
    try:
        return locationDF.loc[(locationDF.CPRO == int(locationCode[0:2])) & (locationDF.CMUN == int(locationCode[2:5]))].values[0][-1]
    except:
        return ''

ddf['location'] = ddf.locationCode.apply(locationLookup)
ddf.head()
```

```python
ddf.to_csv('01_census/data/censusWithLocation.csv', single_file=True)
```
