## 03_xenophobia 

### ESS

```python
# comparing variables across waves
fileColumns = {}
missedFiles = []

for file in glob(f'aux_data/03_xenophobia/ESS/*.csv'):
    
    name = file[file.rfind('\\')+1:-4]
    print(f'doing file {name}')

    try:
        df = pd.read_csv(
            file, 
            nrows=10,
            )

        fileColumns[name] = list(df.columns)

    except UnicodeDecodeError:
        missedFiles.append(name)
        pass

print(f'missed files: {missedFiles}')


locations = {}
count = -1
reorganisedAll = {}


for file, cols in fileColumns.items():
    reorganisedOne = [None] * 3000

    for element in cols:
        if element not in locations.keys():
            count += 1
            locations[element] = count

        reorganisedOne[locations[element]] = element


    reorganisedAll[file] = reorganisedOne





comparisonDF = pd.DataFrame(reorganisedAll)
comparisonDF.to_excel('variable comparison.xlsx')
comparisonDF
```

```python
# stitching all waves together into one database
waves = []
dtypes = {
    'cntbrthc': 'object',
    'ctzshipc': 'object',
    'fbrncntb': 'object',
    'mbrncntb': 'object', 
    'cntbrtha': 'object',
    'ctzshipa': 'object', 
    'cntbrthd': 'object',
    'ctzshipd': 'object',
    'mbrncntc': 'object', 
    'ctzship': 'object',
    }


for file in glob(f'aux_data/03_xenophobia/ESS/*.csv'):
    
    name = file[file.rfind('\\')+1:-4]
    if name == 'all_waves': continue
    print(f'doing file {name}')

    temp = dd.read_csv(
        file, 
        assume_missing=True,
        low_memory=False,
        dtype=dtypes,
        )

    waves.append(temp)

all = dd.multi.concat(
    waves, 
    axis='index', 
    join='outer',
    ignore_index=True,
    )

defragmented = all.copy()

defragmented.to_parquet(
    'aux_data/03_xenophobia/ESS/all_waves/', 
    name_function=lambda x: f'ESS-{x:03}.parquet',
    engine='fastparquet',               
    compression='snappy', 
    write_index=True, 
    append=False, 
    write_metadata_file=True, 
    schema='infer', 
    )
defragmented
```
